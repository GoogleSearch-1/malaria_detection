```python
#Imports
import os
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.datasets import cifar10
from tensorflow.keras import backend as K
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout, BatchNormalization, Normalization
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
from PIL import Image
import cv2
import sys
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Data


images = "/content/drive/MyDrive/malariaDetection/Malaria/cell_images/"
infected = os.listdir(images + 'Parasitized')
uninfected = os.listdir(images + 'Uninfected')
images_train = []
labels_train = []
images_test = []
labels_test = []
x = round(len(infected)*0.8)
for i in range(x):
    image = cv2.imread("/content/drive/MyDrive/malariaDetection/Malaria/cell_images/Parasitized/" + str(infected[i]), cv2.IMREAD_GRAYSCALE)
    try:
        image.resize(128, 128, 1)
    except:
        continue
    images_train.append(image)
    labels_train.append(1)
y =round(len(uninfected)*0.8)
for a in range(y):
    image = cv2.imread("/content/drive/MyDrive/malariaDetection/Malaria/cell_images/Uninfected/"  + str(uninfected[a]), cv2.IMREAD_GRAYSCALE)
    try:
        image.resize(128, 128, 1)
    except:
        continue
    images_train.append(image)
    labels_train.append(0)
for b in range(x, len(infected)):
    image = cv2.imread("/content/drive/MyDrive/malariaDetection/Malaria/cell_images/Parasitized/" + str(infected[b]), cv2.IMREAD_GRAYSCALE)
    try:
        image.resize(128, 128, 1)
    except:
        continue
    images_test.append(image)
    labels_test.append(1)
for c in range(y, len(uninfected)):
    image = cv2.imread("/content/drive/MyDrive/malariaDetection/Malaria/cell_images/Uninfected/" + str(uninfected[c]), cv2.IMREAD_GRAYSCALE)
    try:
        image.resize(128, 128, 1)
    except:
        continue
    images_test.append(image)
    labels_test.append(1)
```

```python
images_train = np.array(images_train, dtype='float32')
images_test = np.array(images_test, dtype='float32')
labels_test = np.array(labels_test, dtype='float32')
labels_train = np.array(labels_train, dtype='float32')
```

```python
#Model
epochs = 100
input_shape = (128, 128, 1)
batch_size = 64
model = Sequential()
model.add(Conv2D(filters=128, kernel_size=(3, 3), activation='relu', input_shape=input_shape, batch_size=batch_size))
model.add(MaxPooling2D())
model.add(Conv2D(64, (3,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3)))
model.add(MaxPooling2D())
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
model.build(input_shape)
model.summary()
```

```python
#Test
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
model.fit(images_train, labels_train, batch_size=batch_size, epochs=epochs, validation_data=(images_test, labels_test), shuffle=True)
scores = model.evaluate(images_test, labels_test,verbose=0)
print('Test accuracy:', str(scores[1]))
```
